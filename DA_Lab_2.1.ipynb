```python
import numpy  as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.cluster import KMeans
from PIL import Image

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('C:/Users/Olga/Downloads/archive/happydata.csv')
print(df)
```

```python
scaler = MinMaxScaler()
df_normalized = scaler.fit_transform(df)

print(df_normalized)
```

```python
X = df.drop('happy', axis=1).copy()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], cmap='viridis', edgecolor='k', s=40)
plt.title('PCA Visualization')
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1],  cmap='viridis', edgecolor='k', s=40)
plt.title('t-SNE Visualization')
```
